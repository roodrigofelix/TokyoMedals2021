```python
plt.figure(figsize=(20, 30))
sns.barplot(x='Total',y='Country',data=df)
plt.title('Rank dos paises por número de medalhas')
plt.show()
```

```python
plt.figure(figsize=(20, 15))
sns.barplot(x='Gold Medal',y='Country',color = '#D1B000',data=df[:15])
plt.title('Distribuição das medalhas de Ouro')
plt.show()
```

```python
plt.figure(figsize=(20, 15))
sns.barplot(x='Silver Medal',y='Country',color='#828A95',data=df[:15])
plt.title('Distribuição das medalhas de Prata')
plt.show()
```

```python
plt.figure(figsize=(20, 15))
sns.barplot(x='Bronze Medal',y='Country',color = '#914E24',data=df[:15])
plt.title('Distribuição das medalhas de Bronze')
plt.show()
```

```python
top_10_countries = df['Country'][:10]
top_10_total = df['Total'][:10]


colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

figure = plt.figure(figsize=(10,10))
plt.pie(top_10_total, labels=top_10_countries,
        colors = colors, startangle = 90, autopct='%1.2f%%')

plt.title('Porcentagem das medalhas conquistadas na competição')
plt.tight_layout()
plt.show()
```

```python
df_brazil = df.loc[df['Country']=='Brazil']
df_brazil
```

```python
fig,ax = plt.subplots(figsize = (10,5))

indices = 1
ind = np.arange(indices) 

width = 1
ax.bar(ind, df_brazil['Gold Medal'], data=df_brazil, width=width, color = '#C49133', label = 'Gold')
ax.bar(ind+width, df_brazil['Silver Medal'], data=df_brazil, width=width, color = '#828A95', label = 'Silver')
ax.bar(ind+2*width, df_brazil['Bronze Medal'], data=df_brazil, width=width,  color = '#914E24', label = 'Bronze')


plt.title('Medals won by Brazil in Tokyo Olympics', fontweight = 'bold', fontsize=20)
plt.xlabel('Brazil', fontsize = 10, fontweight = 'bold')
plt.ylabel('Nº of Medals', fontsize = 10, fontweight = 'bold')
plt.legend(fontsize = 10)


plt.tick_params(axis='x', labelbottom=False)

plt.show()
```
